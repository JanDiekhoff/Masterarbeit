```python
import json

with open("data.json", "r") as file:
    data = json.load(file)

wrong_baseline = []
for row in data:
    if row["base"] != "Hello\nApril 20, 20":
        wrong_baseline.append(row)
        
#print(len(wrong_baseline))

response_types = []
for row in data:
    generations = []
    generations.append(row["base"])
    generations.append(row["direct"])
    generations.append(row["loaded"])
    if row["lora"]:
        generations.append(row["merged"])
        generations.append(row["merged_loaded"])
    
    for gen in generations:
        if gen not in response_types:
            response_types.append(gen)

#print(response_types)

response_counts = {}
for row in data:
    generations = []
    generations.append(row["base"])
    generations.append(row["direct"])
    generations.append(row["loaded"])
    if row["lora"]:
        generations.append(row["merged"])
        generations.append(row["merged_loaded"])
    
    for gen in generations:
        if gen not in response_counts.keys():
            response_counts[gen] = 1
        else:
            response_counts[gen] += 1

response_counts

correct_configs = []
for row in data:
    generations = []
    generations.append(row["base"])
    generations.append(row["direct"])
    generations.append(row["loaded"])
    if row["lora"]:
        generations.append(row["merged"])
        generations.append(row["merged_loaded"])
    
    for gen in generations:
        if gen.find("Banana") >= 0:
            correct_configs.append(row)
            break
        
len(correct_configs)
for c in correct_configs:
    print(c)
```

```python
def count_responses(t):
    response_counts = {}
    for row in data:
        if t in ["merged","merged_loaded"] and not row["lora"]: return
        gen = row[t]
        if gen not in response_counts.keys():
            response_counts[gen] = 1
        else:
            response_counts[gen] += 1
    for c in response_counts:
        print(c, "(", response_counts[c], ")")

for t in ["direct","base","merged","loaded","merged_loaded"]:
    print(t)
    count_responses(t)
    print("\n\n\n")
```
