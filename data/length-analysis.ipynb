```python
import pandas as pd
import datasets
import transformers
```

```python
from datasets import load_dataset

ds = load_dataset("JanDkff/TinyFuncData-docstring", split="train")
```

```python
def add_length_column(example):
    example['body_length'] = len(example['body'])
    example['body_lines'] = len(example['body'].split('\n'))
    return example

dataset = ds.map(add_length_column)
```

```python
import pandas as pd

df = pd.DataFrame(dataset)

lines = df['body_lines']
lengths = df['body_length']
```

```python
import pandas as pd
import numpy as np

min_length = lengths.min()
max_length = lengths.max()
mean_length = lengths.mean()
median_length = lengths.median()
std_length = lengths.std()
percentile_25 = lengths.quantile(0.25)
percentile_75 = lengths.quantile(0.75)

print(f"Minimum length: {min_length}")
print(f"Maximum length: {max_length}")
print(f"Mean length: {mean_length}")
print(f"Median length: {median_length}")
print(f"Standard deviation: {std_length}")
print(f"25th percentile: {percentile_25}")
print(f"75th percentile: {percentile_75}")
```

```python
percentile_90 = lengths.quantile(0.9)
percentile_95 = lengths.quantile(0.95)
percentile_99 = lengths.quantile(0.99)
percentile_999 = lengths.quantile(0.999)
percentile_9999 = lengths.quantile(0.9999)
percentile_99999 = lengths.quantile(0.99999)

print(f"90th percentile cutoff: {percentile_90}")
print(f"95th percentile cutoff: {percentile_95}")
print(f"99th percentile cutoff: {percentile_99}")
print(f"99.9th percentile cutoff: {percentile_999}")
print(f"99.99th percentile cutoff: {percentile_9999}")
print(f"99.999th percentile cutoff: {percentile_99999}")
```
